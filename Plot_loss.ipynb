```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
with open('results.txt', 'r') as f:
    results = np.loadtxt(f, usecols=[2,3,4,5,8,9,10,11,12, 13, 14])
```

```python
train_loss = []
valid_loss = []
for r in results:
    train_loss.append(r[3])
    cur_valid_loss = r[8] + r[9] + r[10]
    valid_loss.append(cur_valid_loss)
```

```python
epoch = [i for i in range(200)]
plt.figure()
plt.plot(epoch, train_loss, color='green', label='train')
plt.plot(epoch, valid_loss, color='blue', label='validation')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.title('Total Loss Curve')
plt.legend()
plt.show()
```
